```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer as Summarizer
from sumy.nlp.stemmers import Stemmer
from sumy.utils import get_stop_words
```

```python
LANGUAGE = "english"
SENTENCES_COUNT = 2
```

```python
def summarize(text):
    parser = PlaintextParser.from_string(text,Tokenizer(LANGUAGE))
    stemmer = Stemmer(LANGUAGE)
    summarizer = Summarizer(stemmer)
    summarizer.stop_words = get_stop_words(LANGUAGE)

    final_sentence = ""
    for sentence in summarizer(parser.document, SENTENCES_COUNT):
        final_sentence += str(sentence)
    return final_sentence
```

## Pin's method

```python
import string
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
```

```python
#nltk.download('words')
#nltk.download('averaged_perceptron_tagger')
```

```python
# remove punctuation
def rev_punc(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation,'')
    return text

# lower case
def cre_low(text):
    word_tokens = word_tokenize(text)
    text = [w.lower() for w in word_tokens]
    return ' '.join(text)

# remove number
def rev_num(text):
    word_tokens = word_tokenize(text)
    text = [w for w in word_tokens if not w.isdigit()]
    return ' '.join(text)

# remove stooword
with open('english_stopwords.txt', 'r') as f:
    stop_words = [line.strip() for line in f]

def rev_stop(text):
    word_tokens = word_tokenize(text)
    text = [w for w in word_tokens if not w in stop_words]
    return ' '.join(text)

#keep noun
def keep_noun(text):
    word_tokens = word_tokenize(text)
    text = [w for (w, pos) in nltk.pos_tag(word_tokens) if pos == 'NN' or pos == 'JJ']
    return ' '.join(text)

# stem
stemmer = SnowballStemmer("english")
def word_lem(text):
    word_tokens = word_tokenize(text)
    text = [stemmer.stem(w) for w in word_tokens]
    return ' '.join(text)


words = set(nltk.corpus.words.words())
#remove english
def rev_nonen(text):
    word_tokens = word_tokenize(text)
    text = [w for w in word_tokens if w in words]
    return ' '.join(text)
```

```python
test_df["summarize"] = test_df["summarize"].apply(rev_punc)
test_df["summarize"] = test_df["summarize"].apply(cre_low)
test_df["summarize"] = test_df["summarize"].apply(rev_num)
test_df["summarize"] = test_df["summarize"].apply(rev_stop)
test_df["summarize"] = test_df["summarize"].apply(keep_noun)
test_df["summarize"] = test_df["summarize"].apply(word_lem)
test_df["summarize"] = test_df["summarize"].apply(rev_nonen)
```

## My method

- remove punctuation

```python
import string
```

```python
def remov_punc(data,column):
    for each in string.punctuation:
        data[column] = data[column].apply(lambda x: x.replace(each,' '))
    data[column] = data[column].apply(lambda x: x.replace("\n",' '))
    return data
```

```python
cleaning_test_df = remov_punc(test_df,"content")
cleaning_test_df = remov_punc(cleaning_test_df,"summarize")
cleaning_test_df = remov_punc(cleaning_test_df,"topic")
```

- *Stop word*

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
def remove_stopword(data,column):
    with open('english_stopwords.txt', 'r') as f:
        stop_words = [line.strip() for line in f]
    index = 0
    for each_words in data[column].apply(lambda x: word_tokenize(x)):
        words = [w for w in each_words if not w in stop_words]
        data.loc[index,column] = " ".join(words)
        index += 1
    return data
```

```python
cleaning_test_df = remove_stopword(cleaning_test_df,"summarize")
cleaning_test_df = remove_stopword(cleaning_test_df,"content")
```

- *Lemmatize*

```python
from nltk.stem import WordNetLemmatizer
```

```python
def lemmatize_word(data,column):
    index = 0
    lemmatizer = WordNetLemmatizer()
    for each_text in data[column].apply(lambda x: word_tokenize(x)):
        words = [w for w in each_text]
        lemmatized = [lemmatizer.lemmatize(word) for word in words]
        data.loc[index,column] = " ".join(lemmatized)
        index += 1
    return data
```

```python
cleaned_data = lemmatize_word(cleaning_test_df,"content")
cleaned_data = lemmatize_word(cleaned_data,"summarize")
```

```python
def create_word_list(X):
    sentences = []
    for each_text in X:
        sentence = []
        for each_word in each_text:
            sentence.append(each_word)
        sentences.append(sentence)
    return sentences
```

## Calculate the vector which represent sentence and topic

```python
import numpy as np
```

```python
def compute_mean_vector(word_list):
    words = [word for word in word_list if word in model.wv.vocab]
    if len(words)>=1:
        return np.mean(model.wv[words], axis=0)
    else:
        return []
```

```python
summarize_vector_list = []
for each_sentence in data_summarize:
    sentence_vector = compute_mean_vector(each_sentence)
    summarize_vector_list.append(sentence_vector)
```

```python
topics_vector_list = []
for each_sentence in data_topic:
    sentence_vector = compute_mean_vector(each_sentence)
    topics_vector_list.append(sentence_vector)
```

# Calculate Cosine Similarity

```python
#!pip install sklearn
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
for i in range(0,len(topics_vector_list)):
    each_summary_vector = summarize_vector_list[i]
    each_topic_vector = topics_vector_list[i]
    if len(each_summary_vector) > 0 and len(each_topic_vector) > 0:
        each_summary_vector =  each_summary_vector.reshape(1,-1)
        each_topic_vector = each_topic_vector.reshape(1,-1)
        similar_rate = cosine_similarity(each_summary_vector,each_topic_vector)
        cleaned_data.loc[i,"similarity_rate"] = similar_rate[0]
    else:
        cleaned_data.loc[i,"similarity_rate"] = 0
#     print(similar_rate)
```

# Calculate accuracy

```python
def calculate_accuracy(data, column1, column2, threshold):
    for i in range(0,len(data)):
        if data.loc[i, column1] > threshold:
            data.loc[i, column2] = 1
        else:
            data.loc[i, column2] = 0
    return data
```

```python
cleaned_data = calculate_accuracy(cleaned_data, "similarity_rate", "similarity", 0.6)
```
